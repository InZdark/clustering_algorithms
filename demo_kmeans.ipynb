```python
from cluster_algorithm.kmeans import Kmeans
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

```python
# create data
cluster1 = np.random.normal(loc=0, scale=1, size=(2000, 2))
cluster2 = np.random.normal(loc=5, scale=1, size=(2000, 2))
x = np.vstack((cluster1, cluster2))
# plot data points
plt.figure(figsize=(7, 7))
plt.scatter(x[:,0], x[:,1], s=5)
plt.show()
```

```python
# use Kmeans for clustering
kmeans = Kmeans(k=2)
# c: centroids
c = kmeans.fit(x)
# y: class
y = kmeans.predict(x)
# plot clusters
plt.figure(figsize=(7, 7))
plt.scatter(x[:,0], x[:,1], cmap='YlOrRd', c=y, s=5, label='data points')
plt.scatter(c[:, 0], c[:, 1], c='orange', label='centroids')
plt.legend()
plt.show()
```
